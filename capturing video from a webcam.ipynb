```python
import cv2

cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("Error: Could not open camera.")
    exit()

fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.avi', fourcc, 20.0, (640, 480))
print("Camera opened:", cap.isOpened())

while True:
    ret, frame = cap.read()
    if not ret:
        print("Error: Failed to capture frame.")
        break  # Exit if frame reading fails

    # Print current frame dimensions (optional)
    print("Width:", cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    print("Height:", cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

    # Write the original frame to output file
    out.write(frame)

    # Convert to grayscale (correct conversion for normal webcams)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cv2.imshow('frame', gray)

    # Exit on 'q' key press (corrected typo: waitKey)
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```
